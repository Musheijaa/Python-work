```python

#check for non numeric values in numeric columns


import pandas as pd

# Read the CSV file
df = pd.read_csv("data.csv")

# Check for non-numeric values in numeric columns
for col in ['Duration', 'Pulse', 'Maxpulse', 'Calories']:
    # Use to_numeric with errors='coerce' to identify bad entries
    non_numeric = df[pd.to_numeric(df[col], errors='coerce').isna()]
    
    if not non_numeric.empty:
        print(f"\nNon-numeric values found in column '{col}':")
        print(non_numeric[[col]])
    else:
        print(f"\nAll values in column '{col}' are numeric.")
```

```python
#wrong data
#check for negative values in numeric columns

import pandas as pd

# Read the CSV file
df = pd.read_csv("data.csv")

# Define the numeric columns to check
numeric_columns = ['Duration', 'Pulse', 'Maxpulse', 'Calories']

# Check for negative values
for col in numeric_columns:
    negative_values = df[df[col] < 0]
    if not negative_values.empty:
        print(f"\nNegative values found in column '{col}':")
        print(negative_values[[col]])
    else:
        print(f"\nNo negative values found in column '{col}'.")
```

```python
#Duplicates
import pandas as pd

# Load your dataset
df = pd.read_csv("data.csv")

# Check for duplicate rows
duplicates = df[df.duplicated()]
print("Duplicate rows:")
print(duplicates)

# Remove duplicates (optional)
df_cleaned = df.drop_duplicates()

# Save cleaned data (optional)
df_cleaned.to_csv("cleaned_data.csv", index=False)
```

```python
#recommendations
#1.Handle empty values
#2.convert wrong formarts
#3.correct wrong data
#4.remove duplicates in rows not in columns


#fill in empty values


import pandas as pd

# Load the dataset
df = pd.read_csv("data.csv")

# Fill all missing values with 0
df.fillna(0, inplace=True)

# OR fill missing values in specific column(s)
df['Calories'].fillna(df['Calories'].mean(), inplace=True)  # Example: use mean for numeric column


# Save the cleaned data (optional)
df.to_csv("cleaned_data.csv", index=False)
```

```python
#Data visualisation the quickest way to make sense out of your  data
import matplotlib.pyplot  as plt
import seaborn as sns




```

```python
#Histogram of 'Duration'
#Set up the figure size
plt.figure(figsize=(15,10))

#Histogram of duration
plt.subplot(2,2,1)
sns.histplot(df['Duration'],bins=30,kde = True)
plt.title("Duration Distribution")
#Histogram of 'Pulse'
plt.subplot(2,2,2)
sns.histplot(df['Pulse'],bins=30,kde=True)
plt.title('Pulse Distribution')
```

```python
#Scatter plot of 'Duration' vs 'Duration'
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.scatterplot(x='Duration',y='Pulse',data=df)
plt.title("Duration vs Pulse")
```

```python
#Collelation heatmap
corr = df.corr()
sns.heatmap(corr,annot = True,cmap = 'coolwarm',fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('data.csv')

# Plot histograms
plt.figure(figsize=(12, 5))

# Maxpulse Histogram
plt.subplot(1, 2, 1)
plt.hist(df['Maxpulse'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Maxpulse')
plt.xlabel('Maxpulse')
plt.ylabel('Frequency')

# Calories Histogram
plt.subplot(1, 2, 2)
plt.hist(df['Calories'].dropna(), bins=20, color='lightgreen', edgecolor='black')
plt.title('Histogram of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

```
